```python
#Extraemos el CSV del archivo zip
import zipfile as zf
zip_ref = zf.ZipFile('../Data/amazon_prime_titles.csv.zip')
zip_ref.extractall('../Data')
zip_ref.close()
```

```python
#Cargamos el archivo CSV en el dataframe y mostramos el dataframe
import pandas as pd
streaming_df = pd.read_csv('../Data/amazon_prime_titles.csv.zip')
streaming_df
```

```python
#Cargamos el dataframe en la base de datos
import sqlalchemy as sql
from dotenv import load_dotenv
import os
```

```python
#Cargamos las variables de entorno para la conexión con la base de datos
load_dotenv()
name_db = os.getenv('DB_NAME')
user_db = os.getenv('DB_USER')
password_db = os.getenv('DB_PASSWORD')
host_db = os.getenv('DB_HOST')
port_db = os.getenv('DB_PORT')
```

```python
#Creamos la conexión con la base de datos
try:
    engine = sql.create_engine(f'mssql+pyodbc://{user_db}:{password_db}@{host_db}/{name_db}?driver=ODBC+Driver+17+for+SQL+Server')
    connection = engine.connect()
    print(f'Conexion exitosa a la base de datos: {name_db}')
except Exception as e:
    print(f'Error al conectar con la base de datos: {e}')
```

```python
#Cargamos el dataframe en la base de datos 
streaming_df.to_sql('amazon_prime_db', con= connection, index= False, if_exists= 'replace')
connection.close()
```
